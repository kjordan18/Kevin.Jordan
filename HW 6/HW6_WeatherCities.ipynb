```python
# Dependencies
import csv
import json
import matplotlib.pyplot as plt
from random import *
import pprint
import requests
import pandas as pd
from config import api_key
from citipy import citipy

# Get our api_key
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
queryUrl = f"{url}appid={api_key}&units={units}&q="
```

```python
# create empty lists for cities and their country code
cities = []
countryCode = []

#conditional while cities is less than 1000
while len(cities) <= 1000:
    # randomly select latitude and longitude
    lat = randint(-90.00, 90.00)
    lon = randint(-180.00, 180.00)
    #find nearest city to the random lat and lon
    city = citipy.nearest_city(lat, lon)
    #if not a repeat city, then append to lists
    if city not in cities:
        cities.append(city.city_name)
        countryCode.append(city.country_code)
    else:
        continue
```

```python

# create empty lists for categories I need
names = []
lat = []
temp = []
cloud = []
wind = []
humidity = []
country = []

# Loop through the list of cities and perform a request for data on each

for city in cities:
    response = requests.get(queryUrl + city).json()
    #print(json.dumps(response, indent=4, sort_keys=True))
    try:
        lat_result = response['coord']['lat']
        #print(lat_result)
        #print(response.keys())
        names.append(city)
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        wind.append(response['wind']['speed'])
        cloud.append(response['clouds']['all'])
        country.append(response['sys']['country'])
    except:
        pass

len(names)

# print(f"The latitude information received is: {lat}")
# print(f"The temperature information received is: {temp}")
# print(f"The humidity information received is: {humidity}")
# print(f"The cloudiness information received is: {cloud}")
# print(f"The wind speed information received is: {wind}")

```

```python
# create a data frame from cities, lat, and temp
citiesDf = pd.DataFrame({'Cities': names,
                         'Latitude': lat,
                         'Temperature (Celsius)': temp,
                         'Humidity': humidity, 
                         'Cloudiness': cloud,
                         'Wind Speed': wind})
citiesDf.head()
```

```python
# Create a scatter plot for each data type
plt.scatter(citiesDf["Latitude"], citiesDf["Temperature (Celsius)"], marker="o")

# Incorporate the other graph properties: title and axis labels
plt.title("Temperature in World Cities")
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Latitude")

# Turn on the grid
plt.grid(True)

# Save the figure
plt.savefig('Temp.png')

# Display to screen 
plt.show()
```
